```python
import pandas as pd
import plotly.express as px
```

```python
#NSANSS0002 - NSW
NSANSS_latlong = 148.32397997,-32.83522996
path_NSANSS0002 = 'C:/Projects/130-Enhancing_FC/hyperspec-fc/field_spectra/data/soil/New South Wales/NSANSS/NSANSS0002/TERN_2019_03_04_merged.csv'

# Read the CSV file with space as the delimiter
df_NSANSS0002 = pd.read_csv(path_NSANSS0002, delim_whitespace=True)

# Display the first few rows of the dataframe
df_NSANSS0002.head()
```

```python
# Melt the dataframe to have 'Wavelength' and 'Reflectance' in long format
df_melted = df_NSANSS0002.melt(id_vars=['Wavelength'], var_name='Sample', value_name='Reflectance')

# Create the plot
fig = px.line(df_melted, x='Wavelength', y='Reflectance', color='Sample', title=' NSANSS0002 Soil Spectra')
fig.show()
```
